```python
# Se requiere upgradear el sistema de adquisición a adquisición + decisión => Velocidad del algoritmo fundamental.

import numpy as np
import matplotlib.pyplot as plt
source="/media/ids-pc2/TOSHIBA EXT/Primavera 2017/Simple-Auto/Pi Manzanas/DCA_160506_LIN/Spectra/0001_CA16.txt"
file = open(source, "r")
# Extraer los datos
dato=list()
wv=list()
values=list()
for line in file:
    dato=(line.split())
    wv.append(float(dato[0].replace(',','.')))
    values.append(float(dato[1].replace(',','.')))
    
wv_int=np.asarray(wv,dtype=int)
values=np.asarray(values)
plt.figure()
plt.plot(wv,values)
plt.show()

# Down-Samplear considerando el promedio
Prom=np.asarray([0])
Prom[0]=values[0]
print(Prom.size)
j=1
for i in range(1,wv_int.size):
    if wv_int[i]==wv_int[i-1]:
        Prom[-1]=np.add(Prom[-1]*j,values[i])/(j+1)
        j=j+1
    else:
        j=1
        Prom=np.append(Prom,values[i])
plt.figure()
plt.plot(Prom)
plt.show()

print(np.unique(wv_int))
    
    
    

    



```
